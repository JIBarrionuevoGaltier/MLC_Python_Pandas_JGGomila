```python
# Para saltar las primeras filas 
data = pd.read_csv(filepath_or_buffer="../datasets/titanic/titanic3.csv",
                   sep=",", dtype={'body': np.float64}, header=0, #names={'pclass': '2'}
                   skiprows=12
                  )
```

```python
# index_col
data = pd.read_csv(filepath_or_buffer="../datasets/titanic/titanic3.csv",
                   sep=",", dtype={'body': np.float64}, header=0, #names={'pclass': '2'}
                   skiprows=12, index_col=1
                  )
```

```python
# saltar valores en blancos
data = pd.read_csv(filepath_or_buffer="../datasets/titanic/titanic3.csv",
                   sep=",", dtype={'body': np.float64}, header=0, #names={'pclass': '2'}
                   #skiprows=12, index_col=1
                   skip_blank_lines=True
                  )
```

```python
data.head()
```

```python
# no tomar NA
data = pd.read_csv(filepath_or_buffer="../datasets/titanic/titanic3.csv",
                   sep=",", dtype={'body': np.float64}, header=0, #names={'pclass': '2'}
                   #skiprows=12, index_col=1
                   skip_blank_lines=True, na_filter=True
                  )
```
